```python
# Date: 4/21/2019
# Course: DSC 510 - Introduction to Programming
# Desc: creates a list and the user inputs temperatures until they are finished. 
# Usage: User enters temperatures until finished, once finished they will enter 'q' to quit and the program will return the min, max and number of temperatures entered. 

def getTemperatureMeasurement():
    #To keep the loop from going into an infite loop I give the user 5 tries to enter a standard measurement
    for i in range(5):
        #prompt the user for meausrement type
        tempType = input('What standard of measurement for temperature will be entered [C]elsius or [F]ahrenheit:')
        #make sure temperature is in Celsius or Fahrenheit, but only pull the first letter to check
        if tempType[0:1].upper() in ['C','F']:
            return tempType[0:1].upper()
        else:
            continue
    #If a valid measurment is not entered then default to Celsius
    return 'C'



#this function asks the user for a temperature and returns the number when finished
def promptForTemperature():  
    #To keep the loop from going into an infite loop I give the user 5 tries to enter a valid number
    for i in range(5):
        #ask the user for a number, we use try to make sure the user enters a number
        try:
            print(f'Please enter a Temperature or press \'q\' to stop:')
            response = input()
            #if q is entered return
            if response[0:1].lower() == 'q':
                return response[0:1].lower()
            #convert to float and return
            return float(response)
        except ValueError:
            #flag the error and let the user know why
            print('ValueError: could not convert string to Temperature, please try again.')
            continue
    #if the user does not enter a valid number the program will end
    return 'q'

            
def getTemperatures():
    tempList = []                              #initializing temperature list
    placeholder = 'Start'                      #initialize the placeholder for getting temperutrers
    while (placeholder != 'q'):                #Conitue until user asks to quit
        placeholder = promptForTemperature()   #Ask for temperature from user
        if placeholder == 'q':                 #Exit loop when user hits q
            break
        tempList.append(float(placeholder))    #add temperature to list
    return tempList

def printDescriptiveData(tempList, tempType):
    print('\n'*2)
    print(f'The Max temperature from your list is: {max(tempList)} {tempType}')      #finds and prints the max temperature entered
    print(f'The Minimum temperature from your list is: {min(tempList)} {tempType}')  #finds the min temperature entered
    print(f'The number of temperatures entered: {len(tempList)}')                    #counts the number of temperatures in list
    
def main():
    temperatures=[]                                 #initializing temperature list
    tempType = getTemperatureMeasurement()          #get the standard measurement being entered by the user
    temperatures = getTemperatures()                #gather temperature from the user
    printDescriptiveData(temperatures, tempType)    #print min, max and count for user to see
                

if __name__ == "__main__":
    main()
```
