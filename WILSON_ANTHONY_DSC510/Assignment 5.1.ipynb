```python
# Date: 4/14/2019
# Course: DSC 510 - Introduction to Programming
# Desc: Does basic math operations, including addition, subtraction, multiplication, division, and average 
# Usage: Based on the user input, it will add, subtract, multiply, divide or calculate the average of a list of numbers provided by the end user. It will continue until the end user lets the program know to stop.


#this function asks the user if they want to continue the program and returns true or false
def getResponse():
    #ask the user if they would like to continue
    response = input('Would you like to continue?(y/n)')
    #Give the user 5 times to give the correct repsonse and then end the program
    for i in range(5):
        #use lower in case the end user capitalizes the Y
        if (response.lower() == "y"):
            return True
        #use lower in case the end user capitalizes the N
        elif (response.lower() == "n"):
            return False
        #If the user responds with something other than y/n than iterate the answer needs to be y or n. 
        else:
            response = input('Please enter \"y\" or \"n\" to continue.')
            continue

#This function asks the user for the math operation to be used and returns the operation
def getMathOperation():
    #create a list of math operations for the user to use, so that we make sure they use one of the four
    ops_list = ['+', '-', '*', '/', 'avg']
    #used to iterate while loop until a correct operations is choose
    isContinue = True
    #ask the user the type of operation to perform, no need to repeat this portion
    print('What type of math operation would you like to perform?')
    while (isContinue == True):
        #ask the user which operation to perform
        print('Enter one of the following:')
        print('\t+ = Addition')
        print('\t- = Subraction')
        print('\t* = Multiplication')
        print('\t/ = Division')
        print('\tavg = Average')
        response = input()
        #if the response is one of the following (+,-,*, /)  return the value otherwise repeat the question
        if response in ops_list:
            return response
        else:
            continue

#this function asks the user for a number and returns the number when finished
def getNumber():
    #used to keep the loop going until the user enters a correct value
    isContinue = True
    while (isContinue == True):
        #ask the user for a number, we use try to make sure the user enters a number
        try:
            print('Please enter a number:')
            #convert reposnse to float if it fails flag the exception otherwise return the value
            response = float(input())
            return response
        except ValueError:
            #flag the error and let the user know why
            print('ValueError: could not convert string to Float.')
            continue
            
#this function will take the number of inputs provided by the user and calculate an average
def calculateAverage():
    #describe function use for the user
    print('This function calculates the average for a list of numbers that are entered.')
    #initialize variables, n = 0 is used for while loop to keep it going until it is greater than 0
    n = 0 #used to define the number of numbers to gather
    #i = 0 #used as iterator to gather numbers in loop
    listNumbers = [] #creates list of numbers for average
    sumNumbers = 0.0 #used to sum the numbers up
    
    print('First, we need to decide on how many numbers to enter for the average calculation.')
    while (n <= 0):
        #use get number and convert it to an integer, no reasons to do this twice
        n = int(getNumber())
        #reprompt user when n is less than 0
        if (n <= 0):
            print('This number is used to determine how many numbers to enter, it must be greater than 0.')
    #loop through n number of times determined by the user and calculate the sum for the average        
    for i in range(n):
        listNumbers.append(getNumber())  #append numbers to list
        sumNumbers = listNumbers[i] + sumNumbers #sum the numbers up
        i = i + 1 #increment the loop
        print('Great, ')
    #print the numbers used as well as the average
    print('This is the average for the numbers given: ' + str(sumNumbers/n))
    return   

#this function takes the mathOperation paramater and gets two numbers and performs 
# the math operation for the user and returns the value
def performCalculation (mathOperation):
    #get first value or value1
    val1 = getNumber()
    #lets the user see their number they choose and asks for another
    print('Great the number you choose was ' + str(val1) + ' Let\'s pick another one!')
    val2 = getNumber()
    #show value 2
    print('Here is your second number ' + str(val2))
    #when addition is passed, adds value1 and value2 
    if mathOperation == '+':
        calculation = val1 + val2
    #when subtraction is passed, subtracts value1 and value2 
    elif mathOperation == '-':
        calculation = val1 - val2
    #when multiplication is passes, multiplies value1 by value2
    elif mathOperation == '*':
        calculation = val1 * val2
    #when division is passed divides value1 by value2
    elif mathOperation == '/':
        #use the try except when divding by zero
        try:
            calculation = val1/val2
        except ZeroDivisionError:
            #in case they divide by zero, asks for another value2 until 0 is not choose
            while (val2 == 0):
                print ('ZeroDivisionError: You can\'t divide by 0, please choose another number!')
                val2 = getNumber()
            calculation = val1/val2
            
    print('Here is the calculation for ' + str(val1) + ' ' + str(mathOperation) + ' '+ str(val2) + ' = ' + str(calculation))
    return
                
    

            

def main():
    responseToContinue = True        #used to start the program as well as keep it going
    while (responseToContinue == True):
        #get operation to be used from end user error checking done in function
        operation = getMathOperation()
        #if avg is choose use that
        if (operation == 'avg'):
            calculateAverage()
        #else the other operations have to be choose so no need for if, since error checking is done in the function
        else:
            performCalculation(operation)
        #Get a y/n from the end user in order for the program to continue
        responseToContinue = getResponse()      
                
                
                

if __name__ == "__main__":
    main()
```
