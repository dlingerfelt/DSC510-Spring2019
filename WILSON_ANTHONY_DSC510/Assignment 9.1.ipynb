```python
# Date: 5/11/2019
# Course: DSC 510 - Introduction to Programming
# Desc: creates a list of words used with their count from a file.  
# Usage: Program goes through line by line of a text file, splits out each word counts it and then writes the word 
#       count at the end of the program to a file with the word used the most displayed first.
import string      # this is used to get rid of punctuation when reading in the file


# makes the dictionary look pretty printed out
def pretty_print(wordCount):
    #display the length or total number of words gathered
    sentence = 'Length of the dictionary: {0}'.format(len(wordCount))
    print(sentence)
    #setting up column headers and line across before printing words and their frequencies
    print('\nWord' + 'Count'.rjust(len(sentence)- 5) + '\n' + '_' * len(sentence))
    #loop through values in dictionary in descending order then print them for the end user to see
    for key, value in sorted(wordCount.items(), key = lambda item: item[1], reverse = True):
        printLine = ('{0}' + '{1}'.rjust(len(sentence)-len(key)-2)).format(key, value)
        print(printLine)
        
# makes the dictionary look pretty printed out
def process_file(wordCount, fileToPrint, rowLength):
    with open(fileToPrint,"a") as f:
        #setting up column headers and line across before printing words and their frequencies
        f.write( '\nWord' + 'Count'.rjust(rowLength- 5) + '\n' + '_' * rowLength + '\n')
        #loop through values in dictionary in descending order then print them for the end user to see
        for key, value in sorted(wordCount.items(), key = lambda item: item[1], reverse = True):
            printLine = ('{0}' + '{1}\n'.rjust(rowLength-len(key)-2)).format(key, value)
            f.write(printLine)
        

#add word to dictionary
def add_words(word,wordCount):
    #when word is not in the dictionary add it with frequency = 1
    if word not in wordCount:
        wordCount[word] = 1
    #when word already exists increase frequency by 1
    else:
        wordCount[word] += 1
            
#process a line from text, which includes getting rid of extra spaces and punctuation, and lower casing everything
def process_line(lineToProcess, wordCount):
    #get rid of extra spaces
    lineToProcess = lineToProcess.rstrip()
    #get rid of punctuation
    lineToProcess = lineToProcess.translate(lineToProcess.maketrans('', '' , string.punctuation))
    #lower case everythign
    lineToProcess = lineToProcess.lower()
    #split each line up and create new variable words
    words = lineToProcess.split()
    #loop through new variable words and pull each word out, send it to add_words to add/increase freq in dictionary for each word
    for word in words:
        add_words(word, wordCount)

#prompts the user for a fileName
def getFileName():
    #Give the user 5 tries to enter an acceptable filename
    n = 0 
    for n in range(5):
        try: 
            #ask for file name
            prompt = "Please enter a the filename to write to:"
            fileName = input(prompt)
            #if the user doesn't enter txt program will do it automatically
            if fileName.find("txt") == -1:
                fileName = fileName + ".txt"
            return fileName
        except:
            print("Please try again!")
    return -1
    
#main function
def main():
    #initialize list
    wordCount = {}
    #get fileName from the user
    fileToPrint = getFileName()
    #if user doesn't enter an acceptable filename exit the program
    if fileToPrint == -1:
        print("Terminating the program, an acceptable filename was not entered.")
        quit()
    #open file
    gba_file = open('gettysburg.txt', 'r')
    #read in text file line be line, and send it of to process_line to create dictionary
    for line in gba_file:
        process_line(line, wordCount)
    #Write the length or total number of words gathered
    sentence = 'Length of the dictionary: {0}\n'.format(len(wordCount))
    with open(fileToPrint, "w") as f:
        f.write(sentence)
    #print out dictionary that shows frequencies of words from file
    process_file(wordCount, fileToPrint, len(sentence))
    #let the user know their file has completed
    notification = "Your file \"{0}\" has been been saved.".format(fileToPrint)
    print(notification)
        

if __name__ == "__main__":
    main()
```
