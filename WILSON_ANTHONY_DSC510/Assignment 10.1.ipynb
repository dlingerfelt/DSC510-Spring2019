```python
# Date: 5/19/2019
# Course: DSC 510 - Introduction to Programming
# Desc: gets Chuck Norris jokes from web api, and returns as the user wishes
# Usage: Program displays instructions on how to interact with the program and returns a joke to the user as long as they
#        want the jokes to keep coming.

import  requests, json, string, random
from IPython.display import clear_output

#pulls data from website and puts in in a dictionary and returns it back
def getDataFromWeb(url):
    #error handling the website, in case it doesn't work
    try:
        #get data
        response = requests.get(url)
        #insert json data into data
        data = response.json()
    except:
        #if the website is down or sends bad data flag an error
        print('Chuck Norris web jokes are down from a complete round house, from you know who! Please try again!')
        #insert an empty set
        data = None
    return data

#get a response from the user whether they want to continue
def getUserResponse(isFirstJoke):
    # used to keep loop for response going
    isContinue = True
    
    #putting list of correct responses together
    validResponses =('y','n')
    
    #basePrompt tells the user how to continue
    basePrompt = "\nPlease enter y-yes, keep the jokes coming or n-no more jokes, I need to catch my breath."
    
    #promptList is a list of questions, asking the user to continue
    promptList = ["Chuck Norris wants you to continue, but the choice is yours.",
                    "Choosing to stop is lethal, but you have to decide for yourself, which will it be?", 
                    "If you choose to stop, it could make Chuck Norris cry tears of molten lava, what are you going to choose?"]
    
    #errorpromtList is a list of responses for the user if the do not enter y/n. 
    errorPromptList = ["You made a mistake by entering that, I believe they call those errors. \nChuck Norris doesn't make those!"
                       "Well that's embarrassing, Chuck Norris saw that mistake and laughed at the error you made!", 
                       "Chuck Norris wants you to enter \"y\" to continue!\nBelieve me he knows the error you made!"]
    
    #we need to welcome the user first, then after that give a prompt that is generic to continue 
    if isFirstJoke:
        intialPrompt = "Welcome, this is going to be great, a few guidelines to follow in order to keep the jokes going:"
        promptUser = "{0} {1}\n".format(intialPrompt, basePrompt)
    else:
        promptUser = "\n{0} {1}\n".format(random.choice(promptList), basePrompt)
    #loop through until the user provides an appropriate response
    while (isContinue == True):
        #error handling the user input
        try:
            userResponse = input(promptUser)
            #make sure the user entered a y or n
            assert userResponse.lower() in validResponses
            #if response is y then send back true to continue else return false
            if (userResponse.lower() == 'y'):
                return True
            else:
                return False
        except:
            #let user know they need to choose y/n
            promptUser = "\n{0} {1}\n".format(random.choice(errorPromptList), basePrompt)

#print the data neatly for the user to read
def prettyPrint(data, n):
    #clears the output so it is easier to read the joke
    clear_output()
    #include the joke number so the user see's what joke they are on
    output = "Joke {0}:  {1}".format(n, data["value"])
    print(output)    
    
def main():
    #website to pull from 
    url = 'https://api.chucknorris.io/jokes/random'
    # used to show the number of iterations the user has gone through
    i = 0
    #used to iterate through the loop to keep jokes going
    isContinue = getUserResponse(True)
    while isContinue:
        #construct the data
        data = {}
        #get the feed back for the data
        data = getDataFromWeb(url)
        prettyPrint(data, i + 1)
        #ask user if they want to continue
        isContinue = getUserResponse(False) 
        i += 1
    #thank the user for coming
    print("\nThanks for coming, please come back soon!")
        
if __name__ == "__main__":
    main()

```
